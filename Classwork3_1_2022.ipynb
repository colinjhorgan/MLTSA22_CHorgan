```python
import pandas as pd
import numpy as np

#supress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/3. MSDS Courses/Machine Learning Time Series/COVID_19/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')
df.isna().sum()
```

```python
df = df.loc[df['state']=='DE']
df.head()
```

```python
cases = df[['submission_date','new_case']]
cases.columns = ['date', 'cases']
cases['date'] = pd.to_datetime(cases['date'])
cases = cases.sort_values(by='date')
cases.head()
```

```python
covid21 = cases.loc[cases['date'].between('2021-01-01', '2021-12-31')]
covid21.plot(x='date', y='cases', figsize=(12,6))
```
